```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train_test.csv')
df
```

```python
stdevs = {}
for col in df.columns[2:]:
    stdevs[col] = [df[col].quantile(0.25), df[col].quantile(0.75)]
stdevs
```

```python
base_df = pd.read_csv('../HackUTD-X-Golden-Hour/sensor_readings.csv')
base_df.columns = df.columns[:26]
stdevs = {}
for col in base_df.columns[2:]:
    stdevs[col] = [base_df[col].quantile(0.25), base_df[col].quantile(0.5), base_df[col].quantile(0.75)]
stdevs
```

```python
stdev_df = pd.DataFrame(stdevs)
stdev_df = stdev_df.transpose()
stdev_df.columns = ['Lower Quartile', 'Mean', 'Upper Quartile']
stdev_df.to_csv('../HackUTD-X-Golden-Hour/stdev.csv')
```
